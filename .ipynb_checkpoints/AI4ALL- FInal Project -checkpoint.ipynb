# What is reinforcement learning??
---
Reinforcement learning is a type of machine learning that revolves around an agent navigating an enviornment/gym taking actions in order to maximize a reward. Reinforcement Learning (RL) concerns issues that are all about optimizing problems (such as which path is the most efficient, or what stock option will yield the most returns) in order to minimize any pentaly (loss) it may recieve as the result of a bad prediction.

In summary when it comes to RL it is best to remember the main concepts with this useful mnemonic, AREA 51 <br>

A- Agent, the AI that is to be trained<br>
R- Reward, what the AI is trying to optimize to get the most amount of<br> 
E- Enviornment, the setting in which the agent is active in<br>
A- Action, the actions an agent can take in an enviornment<br>

# Step 3. Training the Reinforcement Model
---
This step will be all about training the AI to navigate the environment. This uses reinforcement learning with Mario
as the agent. Mario will be rewarded for for how far right he makes it in the enviornment. The goal of Mario is to maximize his reward.

The Reinforcement Alogirithm being used is the PPO (Proximal Policy Optimization) algorithm. 

## PPO (Proximal Policy Optimization)
We now need an algorithm to train the AI, which we will be using one of the most common and proven AI training algorithims PPO (Proximal Policy Optimization)

```python
# Import os for file path management
import os 
# Import PPO for algos
from stable_baselines3 import PPO
# Import Base Callback for saving models to prevent massive data loss 
from stable_baselines3.common.callbacks import BaseCallback
import torch

```

Callback class - This is so that the AI can save it's progress after a certain amount of steps in order to prevent massive data lost

```python
class TrainAndLoggingCallback(BaseCallback):

    def __init__(self, check_freq, save_path, verbose=1):
        super(TrainAndLoggingCallback, self).__init__(verbose)
        self.check_freq = check_freq
        self.save_path = save_path

    def _init_callback(self):
        if self.save_path is not None:
            os.makedirs(self.save_path, exist_ok=True)

    def _on_step(self):
        if self.n_calls % self.check_freq == 0:
            model_path = os.path.join(self.save_path, 'best_model_{}'.format(self.n_calls))
            self.model.save(model_path)

        return True
```

Since it is a good idea to save the model as it changes we will create a directory for the model to be stored at 

```python
CHECKPOINT_DIR = './train/'
LOG_DIR = './logs/'
```

```python
# Setup model saving callback
callback = TrainAndLoggingCallback(check_freq=10000, save_path=CHECKPOINT_DIR)
```

```python
# This is the AI model started
model = PPO('CnnPolicy', env, verbose=1, tensorboard_log=LOG_DIR, learning_rate=0.000001, n_steps=512) 
```

**This is an extremely long process and will take well over a few hours depending on your GPU (AN NVIDIA GPU IS HIGHLY RECOMMENDED FOR TRAINING THE AI TO DECREASE TRAINING TIME) only run the AI training if you wish to build the model yourself, otherwise continue reading**

```python
# Train the AI model, this is where the AI model starts to learn,timesteps is the amount of frames it will be processing
model.learn(total_timesteps=1000000, callback=callback)
```

# Conclusions 
The results of this project can be described as mixed. In the end we were successful in creating an AI that was able to play a game of Super Mario Bros. It was able to progress the stage with no need for human input, and as we can see from it's progression across the three snapshots taken it managed to learn how to successfully overcome various obstales such as pipes and enemies.

Yet we can also observe that the final product after 50K steps of training did hit a snag in being unable to progress past a pipe that was too high for it to simply leap over. This can be pinpointed down to two causes. One, the way Mario is rewarded is by how far to the right he progresses regardless of context. This means that during the training phase Mario will try to move as far to the right as he can as moving left would decrease his reward. What could have been a simply asjustment for a human in that case, to reposition left and attempt the jump again to the AI doing so would have decreased it's reward so therefore it decided not to. If the parameters for the reward were changed then perhaps this specific trained model could have avoided that tendency. 

Another explanation can be seen in how we did not train this AI enough, 50K steps of training simply isn't enough to train an agent especially in a enviornment as complex as the one we have given. Increasing the training phase would have created a stronger model in temrs of performance since it would have had more previous data to work off of and improve.

And lastly a limitation of this project is that we only seen the performance of Mario in a single stage. We do not know how it would have performed in the other stages offered in the Mario-Gym environment and testing the trained model in a multitude of environments could give us a better glimpse for the true strength of this AI.



## Further discussion points: The Ethics of Reinforcement Learning 

This project can serve as a great case study for the limitations and ethical concerns Reinforcement Learning can bring. The shortcommings of this project show a great fault in rushed training or lack of proper testing. When an AI is trained and is using RL as a learning method proper training time is needed, these things simply cannot be rushed (despite how long training may take on some workstations) and when AI is shipped without being trained for long enough like our end results they can be met with subpar performance or at worst create disastrous predictions. For large AI development training may take hundreds or thousands of hours, but when these products are being released to potentially thousands or millions of users these RL AIs need to have been throughly trained to make the best predictions possible.

The second pitfall of RL AIs can be in their testing phase. Gym environments are used to see how the AI will act in a variety of situations which is needed for commercial products as these AIs will be navigating a variety of situations and as such being tested on as many edge cases and possible environments to prevent any unintended predictions or consequences from a bad guess by the AI.
